# Opettelen pyplot:ia

Tutoriaali: https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Eka

```python
# Eka
plt.plot([1,2,3,4,5], [3,5,8,22,66], 'ro')
plt.xlabel('Tama on kuvaaja')
plt.ylabel('Kuikka Koponen')
plt.show()
```

# Toka

```python
plt.plot([2,4,6,8,10], [88,44,28,33,11], [10,20,22,33,44])
plt.xlabel('1 viiva 5')
plt.ylabel('Kekkonen')
plt.text(3,80, r'Tama on tekstia')
plt.annotate('Katsopas tuota', xy=(4,42), xytext=(8,80),
             arrowprops=dict(facecolor='blue', shrink=0.05))
plt.axis([0,12,0,100])
plt.show()

t = np.arange(0., 5., 0.2)
```

# Kolmas: green dashes, red squares and blue triangles

```python
plt.plot(t,t,'g--', t,t**2,'rs', t,t**3, 'b^')
plt.show()
```

# Scatter plot

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

# Category plot

```python
names = ['Eka', 'Toka', 'Kolmas']
values = [20,33,77]

plt.figure(1, figsize=(15,5))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values, linewidth=4.0, alpha=0.5, color='r')
plt.suptitle('Category plots')
plt.show()
```
